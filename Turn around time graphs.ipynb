```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

# Force matplotlib to use a proper backend for saving
matplotlib.use('Agg')  # Non-interactive backend

data = {
    'Partner Type': ['Email', 'LADD', 'RapidILL',  'TPRS'],
    'Avg turnaround (days)': [1.35, 1.58, 0.45, 1.52]
}

df = pd.DataFrame(data)
target = 1.0
df = df.sort_values('Avg turnaround (days)')

fig, ax = plt.subplots(figsize=(10, 4))

# Plot target bars
ax.barh(df['Partner Type'], [target]*len(df), color='lightgrey', height=0.5, label='Target (1 day)')
ax.barh(df['Partner Type'], df['Avg turnaround (days)'], color='#3c1053', height=0.3, label='Actual')
ax.axvline(target, color='red', linestyle='--', label='Target')

# Labels and title
ax.set_title('Digital Borrowing Requests – Turnaround Time vs Target', fontsize=14)
ax.invert_yaxis()
ax.legend()

plt.tight_layout()

# Save figure explicitly
fig.savefig(
    'digital_turnaround_bullet_chart.png',
    dpi=300,
    bbox_inches='tight',
    pad_inches=0.1,
    facecolor='white'  # ensure background is rendered
)

# Optional: show in notebook (interactive rendering)
plt.show()
```
